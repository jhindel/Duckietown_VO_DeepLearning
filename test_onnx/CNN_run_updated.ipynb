```python
# %load_ext autoreload
# %autoreload 2
%matplotlib inline
%pylab inline

import onnxruntime as rt
import onnx

from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from torchvision import transforms

args = {"resize": 64}

# same as for training of model 
transform = transforms.Compose([
            transforms.Resize((args["resize"], args["resize"])),
            transforms.ToTensor(),
            # TODO implement
            # transforms.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))
            # The following means and stds have been pre-computed
            # transforms.Normalize(mean=[0.424, 0.459, 0.218], std=[0.224, 0.215, 0.158])
        ])

def preprocess_img(img):
    # TODO add camera correction here
    area = (0, 160, 640, 480)
    img = img.crop(area)
    img = transform(img)
    return img

# We have to stack two images before to push in the CNN
def preprocess_pose (input_data1,input_data2):
    #convert the 2 input data into float32 input
    img1 = Image.open(input_data1).convert('RGB')
    img2 = Image.open(input_data2).convert('RGB')
    img1 = preprocess_img(img1)
    img2 = preprocess_img(img2)
    newimg = np.concatenate([img1, img2], axis=0)
    return newimg

def display_images(img):
    fig, (ax0, ax1) = plt.subplots((2, 1))
    img1 = np.moveaxis(img[:3], 0, 2)
    img2 = np.moveaxis(img[3:], 0, 2)
    ax0.imshow(np.asarray(img1))
    ax0.set_title('Image 1')
    ax1.imshow(np.asarray(img2))
    ax1.set_title('Image 2')
    fig.show()


img1_name = '/Users/julia/Documents/UNI/Master/Montréal/AV/project/duckietown_visual_odometry/test_onnx/frame000065.png'
img2_name = '/Users/julia/Documents/UNI/Master/Montréal/AV/project/duckietown_visual_odometry/test_onnx/frame000066.png'
input_to_CNN = preprocess_pose(img1_name, img2_name)
display_images(input_to_CNN)

# print(input_to_CNN2.shape)
# input_to_CNN_transpose=np.transpose(input_to_CNN2)
# print(input_to_CNN_transpose.shape)

#Run the CNN model
#session=rt.InferenceSession('2021-12-14-01-11_bestmodel.onnx')
#onnx_model=onnx.load("2021-12-14-01-11_bestmodel.onnx")
#print(onnx.checker.check_model(onnx_model))

model_path = '2021-12-27-16-43_bestmodel.onnx'
onnx_model = onnx.load(model_path)

# Check the model
result=onnx.checker.check_model(onnx_model)
print(result)
#ort_sess=rt.InferenceSession('2021-12-14-01-11_bestmodel.onnx')
#outputs=ort_sess.run(None, {'input': input_to_CNN_transpose})


#put the input to CNN
#input_name=sess.get_inputs()[0].name
#output_name=sess.get_outputs()[0].name
#pred_onx=sess.run(
#    [label_name],{input_name: input_to_CNN_transpose})[0]
#print(pred_onx)
#print(output_name)
    
```
